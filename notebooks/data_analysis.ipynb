**Category**: Tipo de vehículo, como sedan, wagon, etc.

**Model**: Nombre del modelo del vehículo.

**Model Year**: Año del modelo del vehículo, generalmente representado como un número flotante.

**Manufacturer**: Nombre del fabricante del vehículo.

**Fuel**: Tipo de combustible usado por el vehículo (por ejemplo, eléctrico híbrido).

**All-Electric Range**: Rango total en millas que el vehículo puede recorrer usando solo electricidad.

**PHEV Total Range**: Rango total en millas de un vehículo eléctrico híbrido enchufable (PHEV) utilizando ambos modos, eléctrico y de combustión.

**Alternative Fuel Economy City**: Eficiencia de combustible en ciudad usando un combustible alternativo, en millas por galón (MPG).

**Alternative Fuel Economy Highway**: Eficiencia de combustible en carretera usando un combustible alternativo, en millas por galón (MPG).

**Alternative Fuel Economy Combined**: Eficiencia combinada de combustible usando un combustible alternativo, en millas por galón (MPG).

**Conventional Fuel Economy City**: Eficiencia de combustible en ciudad usando combustible convencional, en millas por galón (MPG).

**Conventional Fuel Economy Highway**: Eficiencia de combustible en carretera usando combustible convencional, en millas por galón (MPG).

**Conventional Fuel Economy Combined**: Eficiencia combinada de combustible usando combustible convencional, en millas por galón (MPG).

**Transmission Type**: Tipo de transmisión del vehículo (por ejemplo, automática, manual).

**Transmission Make**: Fabricante de la transmisión.

**Engine Type**: Tipo de motor del vehículo (por ejemplo, motor de inyección directa, turboalimentado).

**Engine Size**: Tamaño del motor en litros.

**Engine Cylinder Count**: Número de cilindros del motor.

**Number of Passengers**: Número de pasajeros que el vehículo puede llevar.

**Heavy-Duty Power System**: Indica si el vehículo tiene un sistema de potencia para trabajo pesado.

**Notes**: Notas adicionales sobre el vehículo.

**Drivetrain**: Tipo de tracción del vehículo (por ejemplo, tracción delantera, tracción total).

### Distribucion de categorias

```python
fig = plt.figure(figsize=(12, 5))
ax = sns.countplot(data=electric_vehicles, 
                   y='Category', 
                   order=electric_vehicles['Category'].value_counts().index, 
                   palette="viridis", 
                   orient='h')
plt.title('Categories distribution')
for p in ax.patches:
    width = p.get_width()
    ax.annotate(f'{int(width)}', (width, p.get_y() + p.get_height() / 2),
                ha='center', va='center', xytext=(5, 0), textcoords='offset points')
plt.show()
```

Las categorias que mas predominan en los datos son **SUV** y **Sedan/Wagon**

### Tipos de combustible

**Hybrid Electric**: Vehículos híbridos eléctricos que combinan un motor de combustión interna con uno o más motores eléctricos y una batería. Pueden recuperar energía mediante frenado regenerativo.

**Plug-in Hybrid Electric**: Similar a los híbridos eléctricos, pero con la capacidad de enchufarse a una fuente externa de electricidad para cargar la batería, permitiendo mayor autonomía eléctrica.

**Electric**: Vehículos totalmente eléctricos que funcionan únicamente con electricidad almacenada en baterías recargables. No tienen motor de combustión interna.

**Biodiesel (B20)**: Vehículos que utilizan una mezcla de 20% biodiesel y 80% diesel convencional. El biodiesel se produce a partir de fuentes renovables como aceites vegetales o grasas animales.

**Ethanol (E85)**: Vehículos que pueden funcionar con una mezcla de 85% etanol y 15% gasolina. El etanol es un biocombustible producido a partir de maíz u otros cultivos.

**Hydrogen Fuel Cell**: Vehículos que utilizan celdas de combustible de hidrógeno para generar electricidad a partir de una reacción química entre el hidrógeno y el oxígeno. Solo emiten agua como subproducto.

**CNG - Bi-fuel**: Vehículos que pueden funcionar tanto con gas natural comprimido (CNG) como con otro tipo de combustible, como gasolina o diesel.

**CNG - Compressed Natural Gas**: Vehículos que utilizan gas natural comprimido como su único combustible. El gas natural es un combustible fósil que produce menos emisiones que la gasolina o el diesel.

**Propane - Bi-fuel**: Vehículos que pueden funcionar tanto con propano (autogas) como con otro tipo de combustible, como gasolina.

**Propane**: Vehículos que utilizan propano como su único combustible. El propano es un gas licuado de petróleo que se quema más limpio que la gasolina o el diesel.

**LNG - Liquified Natural Gas|CNG - Compressed Natural Gas**: Vehículos que pueden funcionar tanto con gas natural licuado (LNG) como con gas natural comprimido (CNG).

**CNG - Compressed Natural Gas|LNG - Liquified Natural Gas**: Vehículos que pueden funcionar tanto con gas natural comprimido (CNG) como con gas natural licuado (LNG).

**Propane|CNG - Compressed Natural Gas**: Vehículos que pueden funcionar tanto con propano como con gas natural comprimido (CNG).

**CNG - Compressed Natural Gas|Plug-in Hybrid Electric**: Vehículos que combinan un sistema de gas natural comprimido (CNG) con la capacidad de funcionar como híbrido enchufable.

**Diesel/Hybrid Electric**: Vehículos que combinan un motor diesel con un sistema de propulsión híbrido eléctrico.

**Propane - Bi-fuel|CNG - Bi-fuel|Propane|CNG - Compressed Natural Gas|Ethanol (E85)**: Vehículos que pueden funcionar con múltiples tipos de combustibles, incluyendo propano, gas natural comprimido, y etanol (E85).

**Propane - Bi-fuel|Propane|CNG - Bi-fuel|Ethanol (E85)|CNG - Compressed Natural Gas**: Vehículos que pueden funcionar con una variedad de combustibles como propano, gas natural comprimido, y etanol (E85).

**CNG - Compressed Natural Gas|Propane - Bi-fuel|E85/Hybrid Electric|CNG - Bi-fuel|Propane**: Vehículos que pueden funcionar con múltiples tipos de combustibles, incluyendo gas natural comprimido, propano, etanol (E85) y sistemas híbridos eléctricos.

**CNG - Bi-fuel|Ethanol (E85)|CNG - Compressed Natural Gas|Propane|Propane - Bi-fuel**: Vehículos que pueden funcionar con una variedad de combustibles, incluyendo gas natural comprimido, propano y etanol (E85).

**Propane - Bi-fuel|Propane|CNG - Compressed Natural Gas|CNG - Bi-fuel**: Vehículos que pueden funcionar tanto con propano como con gas natural comprimido, en sistemas mono y bi-combustible.

**CNG - Compressed Natural Gas|Propane**: Vehículos que pueden funcionar tanto con gas natural comprimido como con propano.

**CNG - Compressed Natural Gas|Propane - Bi-fuel**: Vehículos que pueden funcionar con gas natural comprimido o como bi-combustible utilizando propano.

### Tipos de motor

**SI (Spark Ignition)**:
Definición: Motores de encendido por chispa.
Funcionamiento: Utilizan una bujía para encender la mezcla de aire y combustible en la cámara de combustión.
Ejemplos: Motores de gasolina.

**e-motor (Electric Motor)**:
Definición: Motores eléctricos.
Funcionamiento: Utilizan la electricidad para generar movimiento a través de un campo magnético, sin necesidad de combustión interna.
Ejemplos: Motores en vehículos eléctricos (EV) puros.

**CI (Compression Ignition)**:
Definición: Motores de encendido por compresión.
Funcionamiento: La combustión del combustible ocurre debido a la alta presión y temperatura del aire comprimido, sin necesidad de bujías.
Ejemplos: Motores diésel.

**FC (Fuel Cell)**:
Definición: Celdas de combustible.
Funcionamiento: Generan electricidad a través de una reacción química entre el hidrógeno y el oxígeno, sin combustión directa.
Ejemplos: Vehículos con celdas de combustible de hidrógeno.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(
    x='Transmission Type', 
    y='Conventional Fuel Economy Combined', 
    data=electric_vehicles[['Transmission Type', 'Conventional Fuel Economy Combined']])
plt.xlabel('Tipo de Transmisión')
plt.ylabel('Eficiencia de Combustible Convencional (MPG)')
plt.title('Comparación de Eficiencia de Combustible por Tipo de Transmisión')
plt.xticks(rotation=45)
plt.show()
```

```python
for col in ['Alternative Fuel Economy City', 'Alternative Fuel Economy Highway', 'Alternative Fuel Economy Combined', 'Conventional Fuel Economy City', 'Conventional Fuel Economy Highway', 'Conventional Fuel Economy Combined']:
    plt.figure(figsize=(8,5))
    ax = sns.scatterplot(x='Engine Type', 
                    y=col,
                    data=electric_vehicles,
                    hue='Engine Type')
    ax.set_title(f"Engine Type VS {col}")
    plt.show()
```

En cuanto a la eficiencia de combustible, se observa que usando un **combustible alternativo**, el motor de tipo **e-motor** que corresponde a los vehículos eléctricos, es mas eficiente que el resto de los motores, tanto en ciudad, ruta o combinado.
Respecto al uso de **combustible convencional**, el motor de tipo SI(Spark Ignition) es mas eficiente que el motor de tipo FC(Fuel Cell)
